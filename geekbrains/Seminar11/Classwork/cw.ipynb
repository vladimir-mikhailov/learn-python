1. Определить корни

```python
from decimal import Decimal, getcontext
from sympy import *
getcontext().prec = 20
# from sympy.abc import x
x = Symbol("x", real=True)
f = Function('f')
# Корни
f = collect((5*x ** 2 + 10*x - 30), x)
roots = solve(Eq(f, 0), x)
print(f'Корень 1: {roots[0]}', f'Корень 2: {roots[1]}', sep='\n')
```

2. Найти интервалы, на которых функция возрастает

```python

df = f.diff(x)
solve(df > 0)
```

3. Найти интервалы, на которых функция убывает

```python
solve(df < 0)
```

4. Построить график

```python

from sympy.plotting import plot
plot(f, (x, -10, 10))
```

5. Вычислить низину

```python
df_x = solve(df, x)[0]
print('Y =', f.subs(x, df_x), f' X = {df_x}')
```
